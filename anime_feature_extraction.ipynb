```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"Downloads/anime.csv")
```

```python
df.head() ## top 5 rows
```

```python
## Extracting number of episodes in an anime 
```

```python
def extract_episodes(txt):
    check=False
    data=""
    for i in txt:
        if i==")":
            check=False 
            break
        if i=="(":
            check=True
            continue
        if check==True:
            data+=i
    return data
```

```python
df["Episodes"]=df["Title"].apply(extract_episodes)
df.head()
```

```python
## Converting the whole Episode string  to int
```

```python
def removeeps(txt):
    data=""
    for i in txt:
        if(i==" "):
            break
        data+=i    
    return data  
```

```python
df['Episodes']=df['Episodes'].apply(removeeps)
# or you can do this 
df['Episodes']=df["Episodes"].str.replace(" eps","")
df["Episodes"]=df['Episodes'].astype(int)
df.head()
```

```python
## Extracting Time stamp from the title 
```

```python
def extract_timestamp(txt):
    data=''
    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1,i+1+19):
                data+=txt[j]
            break
    return data
```

```python
df['Time-Stamp']=df["Title"].apply(extract_timestamp)
df.head()
```

```python
## count number of month in each anime
```

```python
def get_months(time_stamp):
    try:
        start, end = time_stamp.split(" - ")
        start = pd.to_datetime(start)
        end = pd.to_datetime(end)
        
        months = (end.year - start.year) * 12 + (end.month - start.month)
        return months
    except:
        return None
```

```python
df['Running-Months']=df['Time-Stamp'].apply(get_months)
df.head()
```
